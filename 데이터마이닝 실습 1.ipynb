```python
df1 = dict()
df1['이름'] = ['김하나', '최금융']
df1['나이'] = [28, 35]
df1['소속'] = ['티아이', '캐피탈']
df1 = pd.DataFrame(df1)

df2 = dict()
df2['이름'] = ['박하나', '이은행']
df2['나이'] = [27, 38]
df2['소속'] = ['손해보험', '저축은행']
df2 = pd.DataFrame(df2)

df_new = pd.concat([df1, df2], axis = 0)
df_new = df_new.reset_index(drop = True)
df_new
```

```python
df_cond = []

df_cond.append(['김하나', 28, '티아이', 2, 'M'])
df_cond.append(['최금융', 35, '캐피탈', 7, 'F'])
df_cond.append(['박그룹', 27, '손해보험', 3, 'M'])
df_cond.append(['이은행', 38, '저축은행', 5, 'M'])
df_cond = pd.DataFrame(df_cond, columns = ['이름', '나이', '소속', '근속연수', '성별'])

cond = df_cond['나이'] < 30
df_new = df_cond.loc[cond]

df_new
```

```python
cond1 = df_cond['소속']!='손해보험'
cond2 = df_cond['성별'] == 'M'
df_new = df_cond.loc[cond1 & cond2]
df_new
```

```python
df_new = df_cond.sort_values('나이', ascending = True)
df_new = df_new.reset_index(drop = True)
df_new
```
