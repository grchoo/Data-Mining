#K means

```python
#K means algorithm
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
plt.figure(figsize = (10, 5))
```

```python
df = pd.read_csv('menu.csv', engine = 'python')
df_data = df[['가격', '판매량']]
km = KMeans(n_init=10, n_clusters = 4, random_state = 0)
km.fit(df_data)
result = km.predict(df_data)
```

```python
km.cluster_centers_
df['Cluster'] = result
```

```python
plt.scatter(df['가격'], df['판매량'], c = df['Cluster'])
plt.show()
```

# 군집화

```python
# DBSCAN

import pandas as pd
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
db = DBSCAN(eps = 10, min_samples = 3)
cluster_pred = db.fit_predict(house_position)
df['CLUSTER'] = cluster_pred
```

```python
x = house_position['x']
y = house_position['y']
plt.scatter(x, y, c = cluster_pred)
```

# Hierarchial Clustering

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/데이터마이닝/2023Fall_데이터마이닝_Python실습II_데이터/housing.csv', engine = 'python')
house_position =df[['x', 'y']]

linked = linkage(house_position, 'single')
dendrogram(linked, orientation = 'top')
```
